 ## probability distribution of data
 * Bernauli Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

p = 0.6
data = bernoulli.rvs(p, size=1000)

plt.bar([0, 1], [sum(data == 0), sum(data == 1)], color=['hotpink', 'lavender'], tick_label=["Failure (0)", "Success (1)"])
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.title("Bernoulli Distribution (p=0.6)")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
mu=70
sigma=10
x=np.linspace(mu-4*sigma,mu+4*sigma,100)
pdf=norm.pdf(x,mu,sigma)
plt.bar(x,pdf,color='purple',label=f"N{mu},{sigma}**2)")
plt.xlabel("Exam Scores")
plt.ylabel("Probability Density")
plt.title("Normal Distribution (={mu},={sigma})")
plt.show()
```
