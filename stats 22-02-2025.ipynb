```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

p = 0.6
data = bernoulli.rvs(p, size=1000)

plt.bar([0, 1], [sum(data == 0), sum(data == 1)], color=['hotpink', 'lavender'], tick_label=["Failure (0)", "Success (1)"])
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.title("Bernoulli Distribution (p=0.6)")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
mu=70
sigma=10
x=np.linspace(mu-4*sigma,mu+4*sigma,100)
pdf=norm.pdf(x,mu,sigma)
plt.bar(x,pdf,color='purple',label=f"N{mu},{sigma}**2)")
plt.xlabel("Exam Scores")
plt.ylabel("Probability Density")
plt.title("Normal Distribution (={mu},={sigma})")
plt.show()
```

```python
z=np.linspace(-4,4,1000)
pdf=norm.pdf(z,0,1)
plt.plot(z,pdf,color='blue',label="Standard normal distribution")
plt.xlabel("Z-scores")
plt.ylabel("Probability density")
plt.title(f"Standard Normal Distribution (Z-distribution)")
plt.axvline(x=0,color='blue',linestyle='--',label="Mean (z=0)")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)

# Generate skewed income data using exponential distribution
income_data = np.random.exponential(scale=50, size=1000)

# Plot original income distribution
plt.figure(figsize=(8, 4))
sns.histplot(income_data, bins=50, kde=True, color='r')
plt.title("Original Income Distribution (Skewed)")
plt.xlabel("Income")   # Added missing xlabel
plt.ylabel("Frequency") # Added missing ylabel
plt.show()

# Sampling distribution of sample means
sample_means = []
n_samples = 1000
sample_size = 30

for _ in range(n_samples):  # Fixed syntax error
    sample = np.random.choice(income_data, size=sample_size)
    sample_means.append(np.mean(sample))

# Plot the sampling distribution of sample means
plt.figure(figsize=(8, 4))
sns.histplot(sample_means, bins=50, kde=True, color="b")  # Fixed missing 'bins'
plt.title("Sampling Distribution of Sample Mean")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```
